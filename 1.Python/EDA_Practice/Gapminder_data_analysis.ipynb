## 연습문제 1

### 문제1
pop 평균보다 인구가 높은 국가의 1970년대 데이터를 필터링하고, continent, country, pop, gdpPercap 열만 출력하라.


```python
popUp1970_9 = data[(data['pop'] > data['pop'].mean()) & (data['year'].isin(list(range(1970,1980))))]
popUp1970_9
```

```python
result = popUp1970_9[['continent', 'country', 'pop', 'gdpPercap']]
result
```

### 문제2
문제1에서 추출한 데이터프레임에서 한국, 중국, 일본의 데이터를 추출하고, 국가별로 그룹화해서 인구수의 평균을 구하라.

```python
# country 명칭 확인
result['country'].unique()
```

```python
# 한, 중, 일 데이터 추출
korea = result[result['country'] == 'Korea, Rep.']
Japan = result[result['country'] == 'Japan']
China = result[result['country'] == 'China']
```

```python
# 각 국가 인구 평균
popMean_kor = korea['pop'].mean()
popMean_jan = Japan['pop'].mean()
popMean_cha = China['pop'].mean()
result_dict = {'Korea, Rep.':popMean_kor, 'Japan':popMean_jan, 'China':popMean_cha}
mean_result = pd.Series(result_dict, dtype=int, name='pop')
# print("population maen of korea : {}\npopulation maen of Japan : {}\npopulation maen of China : {}".format(popMean_kor, popMean_jan, popMean_cha))
mean_result.index.name = 'country'
mean_result
```

```python
# 모범 답안
east_asia=['Korea, Rep.', 'Japan', 'China']

result2 = result[result['country'].isin(east_asia)].groupby(['country']).mean()['pop']
result2
```

## 연습문제 2
* melt, pivot_table 연습

### 문제1
데이터 셋 생성

1. 각각 10명의 학생수를 갖는 두 개의 class A, B를 생성한다.

1. 각 클래스의 학생은 id : 1 ~ 10으로 구분된다.

1. 각 학생들에 대해 math, eng, sci 과목에 대한 30 ~ 100 사이의 random 점수를 생성하라.</br>최종 scores 데이터 프레임은 아래와 같다.

class|stu|math|eng|sci
--|--|--|--|--
A|1|85|92|51
A|2|54|52|42
A|3|89|43|35
...|...|...|...|...
B|9|64|67|62
# B|10|85|89|42

```python
data_A = {
 'class':['A']*10,
 'stu':np.arange(1,11),
 'math':np.random.randint(30,100,10),
 'eng':np.random.randint(30,100,10),
 'sci':np.random.randint(30,100,10)
}
classA = pd.DataFrame(data_A)
classA
```

```python
data_B = {
 'class':['B']*10,
 'stu':np.arange(1,11),
 'math':np.random.randint(30,100,10),
 'eng':np.random.randint(30,100,10),
 'sci':np.random.randint(30,100,10)
}
classB = pd.DataFrame(data_B)
classB
```

```python
class_data = pd.concat([classA, classB])
class_data
```

### 문제2
melt 함수를 이용해서 subj 변수를 새롭게 만들고, 그 변수의 값에 ‘math’, ‘sci’, ‘eng’를 갖게 하고, score라는 변수를 만들어 3개의 과목에 대한 각각의 점수를 갖도록 하라(학생별로 3과목이 연달아 오도록 할 것)

```python
melted_class_data = pd.melt(class_data, id_vars=['class', 'stu'], var_name='subj', value_name='score')
result2 = melted_class_data.sort_values('stu')
result2
```

### 문제3
문제2에서 만든 데이터 프레임을 사용해서 pivot_table을 적용해서 문제1의 데이터 형태로 만들고, 학생별 math, eng, sci의 변수의 평균의 값으로 변수 mean을 생성하라

```python
class_data2 = pd.pivot_table(result2, index=['class','stu'],values='score',columns='subj')
class_data2 = class_data2.reset_index()
class_data2
```

```python
class_data2['mean'] = (class_data2['eng'] + class_data2['math'] + class_data2['sci'])/3
class_data2
```

```python
# class_data2.drop('index',axis=1,inplace=True)
class_data2.columns.name=''
result3 = class_data2
```
